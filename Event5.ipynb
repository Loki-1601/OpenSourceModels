```python
# Create analysis prompt 1
prompt = f"""
**Function:** analyze_grades

The data below contains student grades in Math and Physics for 40 students.
Please provide a simple analysis of this data, including:
1. The average grade for each subject
2. The highest and lowest grades for each subject
3. A brief comparison of performance between Math and Physics

Data: {data}
"""
```

```python
output = llm.invoke(prompt)
print(output)
```

```python
# Prompt 2: Grade Distribution
prompt_grade_distribution = f"""
**Function:** analyze_grade_distribution

The data below contains student grades in Math and Physics for 40 students.
Please analyze the grade distribution for each subject:
1. How many students scored 90 or above?
2. How many students scored between 80 and 89?
3. How many students scored between 70 and 79?
4. How many students scored between 60 and 69?
5. How many students scored below 60?

Provide this information for both Math and Physics.

Data: {data}
"""
output = llm.invoke(prompt_grade_distribution)
print(output)
```

```python
# Prompt 3: Subject Comparison
prompt_subject_comparison = f"""
**Function:** compare_subjects

The data below contains student grades in Math and Physics for 40 students.
Please compare the performance in Math and Physics:
1. In which subject did students perform better on average?
2. Which subject had more students scoring 80 or above?
3. In which subject was the lowest grade recorded?
4. What is the difference between the highest Math grade and the highest Physics grade?

Data: {data}
"""
output = llm.invoke(prompt_subject_comparison)
print(output)
```

```python
# Prompt 4: Identifying Top Performers
prompt_top_performers = f"""
**Function:** identify_top_performers

The data below contains student grades in Math and Physics for 40 students.
Please identify the top performers:
1. How many students scored 90 or above in both subjects?
2. What is the average Math grade for students who scored 90 or above in Physics?
3. What is the average Physics grade for students who scored 90 or above in Math?
4. How many students have a combined grade (Math + Physics) of 180 or higher?

Data: {data}
"""
output = llm.invoke(prompt_top_performers)
print(output)
```
